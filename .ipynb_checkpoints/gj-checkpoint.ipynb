```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/tubes2_HeartDisease_train.csv')
```

```python
df.head()
```

```python
df.replace({"?":np.nan}, inplace=True)
df.isnull().sum(axis = 0)/len(df)
```

```python
df.drop(['Column12', 'Column13'], axis=1)
for column in df.columns:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
df = pd.get_dummies(data=df, columns=['Column3','Column7','Column11'])
```

```python
from sklearn.preprocessing import StandardScaler
cols_to_norm = ['Column4','Column5','Column8','Column10']
df[cols_to_norm] = StandardScaler().fit_transform(df[cols_to_norm])
df.describe()
```

```python
import seaborn as sns

colormap = plt.cm.RdBu
plt.figure(figsize=(14,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(df.astype(float).corr(),linewidths=0.1,vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=True)
```

```python
#df.drop(['Column4', 'Column3_1', 'Column4', 'Column11_2', 'Column7_2'], axis=1)
#exp = df[['Column3_4']]
```

```python
y = df['Column14']
X = df.drop(['Column14'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=123)
```

## Decision Tree

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn import tree
dt = tree.DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
```

```python
accuracy_score(y_test, dt_pred)
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
gnb_pred = gnb.predict(X_test)
```

```python
accuracy_score(y_test, gnb_pred)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
```

```python
accuracy_score(y_test, rf_pred)
```

## kNN

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
train_acc = []
test_acc  = []
for k in range(1,100):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_acc.append(knn.score(X_train ,y_train))
    test_acc.append(knn.score(X_test ,y_test))
```

```python
plt.plot(train_acc)
plt.plot(test_acc)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
```

```python
accuracy_score(y_test, knn_pred)
```

## MLPC

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlpc = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(18,18,18,18), random_state=1, max_iter=1000)
mlpc.fit(X_train, y_train)
mlpc_pred = mlpc.predict(X_test)
```

```python
accuracy_score(y_test, mlpc_pred)
```
